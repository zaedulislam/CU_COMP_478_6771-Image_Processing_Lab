# Read and Display an Image

```python
I = cv2.imread('./Images/cat.png')

# Covert from BGR to GRAY 
J = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)

print('Image size:', I.shape)
print('Image size:', J.shape)

# Color image
cv2_imshow(I)

# Grayscale image
cv2_imshow(J)
```

# Display Image using Matplotlib

```python
I = cv2.imread('./Images/cat.png')

# Covert from BGR to RGB 
Ic = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

print('Image size:', I.shape)
plt.imshow(Ic)

# plt.axis('off')  # to hide tick values on X and Y axis

plt.show()

# Covert from BGR to GRAY
J = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)

print('Image size:', J.shape)
plt.imshow(J, cmap ='gray')
```
